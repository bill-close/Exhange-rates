## Describing the daily exchange rates between 1999 and 2021 for the Euro and USD

### 1. Introduction             
This dataset is available on Kaggle: 
https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020       
    

        

```python
import pandas as pd
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
```

```python
exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

There are 40 columns and 5699 rows to examine.                  
Some of the data has null values. Especially for "Greek drachma" (Perhaps consider deleting this one)      
All of the datatypes are objects with the exception of Iceland krona, Romanian leu and Turkish lira which have float64. 

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
exchange_rates.head()
```

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
euro_to_dollar
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

There are 62 times the character "-" within the dataset

Lets select all the rows where the US_dollar does not contain the "-"

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
```

```python
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
```

```python
euro_to_dollar.info()
```

### 2. Visualising the data     

```python
import matplotlib.pyplot as plt
%matplotlib inline 

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

To smooth out the graph lets calculate the rolling mean

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

This is the new plot. It is visually easier to see

```python
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['rolling_mean'])
plt.xlabel('Year')
plt.ylabel('Exchange rate from 1 Euro to USD')
plt.title('Exhange rate EUR/USD over 20 years')
plt.show()
```

With a rolling average of 30 days applied we can see a more representable graph of the data

### 3. Analysis of exchange rates and the US Presidencies of Bush, Obama and Trump

To do this I will split up the above graph to better address how the exchange was affected in each presidency.        
Bush (2001-2009)       
Obama (2009-2017)          
Trump (2017-2021)        

```python
bush_obama_trump = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2001) & (euro_to_dollar['Time'].dt.year < 2021)]

bush = bush_obama_trump.copy()[bush_obama_trump['Time'].dt.year < 2009]

obama = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year >= 2009) & (bush_obama_trump['Time'].dt.year < 2017)]

trump = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year >= 2017) & (bush_obama_trump['Time'].dt.year < 2021)]

bush
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1,
                                        figsize=(16,11))
axes = [ax1, ax2, ax3]
titles = ['Bush', 'Obama', 'Trump']

for ax in axes:
    ax.plot(bush_obama_trump['Time'], bush_obama_trump['rolling_mean'], color='grey', alpha=0.6)
    for location in ['right', 'top']:
        ax.spines[location].set_visible(False)
        
ax1.text(730799, 1.8, 'Comparison of the EUR/USD exchange under different presidencies', size=18, weight='bold')

ax1.plot(bush['Time'], bush['rolling_mean'], color='chocolate', linewidth='7')
ax1.text(731799, 1.5, 'Bush', size=14, weight='bold', color='chocolate')
ax1.text(729699, 1, 'EUR USD Exchange', size=10, rotation='90')

ax2.plot(obama['Time'], obama['rolling_mean'], color='blue', linewidth='7')
ax2.text(734799, 1.49, 'Obama', size=14, weight='bold', color='blue')
ax2.text(729699, 1, 'EUR USD Exchange', size=10, rotation='90')

ax3.plot(trump['Time'], trump['rolling_mean'], color='red', linewidth='7')
ax3.text(736799, 1.35, 'Trump', size=14, weight='bold', color='red')
ax3.text(729699, 1, 'EUR USD Exchange', size=10, rotation='90')
ax3.text(733999, 0.65, 'Year', size=18)

#plt.grid()      Personally, I didnt like the grid overlay

### Adding a signature
ax3.text(729999, 0.52, 'Bill-Close' + ' '*130 + 'DataSource: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=14)          

plt.show()
```
